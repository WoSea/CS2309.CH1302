```python
from sklearn.datasets import load_iris
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
iris = load_iris()
type(iris)
```

```python
print("Feature names:")
print(iris.feature_names)
print("Data(X):")
print(iris.data[:3])
print("Target names:[0   1   2]")
print(iris.target_names)
print("Target(y):")
print(iris.target)
```

```python
print(type(iris.data))
print(type(iris.target))
print(iris.data.shape)
print(iris.target.shape)
```

```python
X = iris.data
y = iris.target
```

```python
import pandas as pd
data=pd.DataFrame(X,columns=iris.feature_names)
data['target']=pd.DataFrame(y.reshape(-1,1),columns=["target"])
data.head(5)
```

```python
from matplotlib import cm
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(121)
ax.set_title('Iris Classification Based on Sepal length and width', fontsize=20)
ax.set_xlabel('Sepal Length (cm)', fontsize=18)
ax.set_ylabel('Sepal Width (cm)',fontsize=18)
ax.grid(True,linestyle='-',color='0.75')
ax.scatter(data['sepal length (cm)'], data['sepal width (cm)'], s = 100,c=data['target'], marker='o', cmap=cm.jet)
ax = fig.add_subplot(122)
ax.set_title("iris Classification Based on Petal length and width", fontsize=20)
ax.set_xlabel('Petal Length (cm)', fontsize=18)
ax.set_ylabel('Petal width (cm)', fontsize=18)
ax.grid(True, linestyle='-', color='0.75')
ax.scatter(data['petal length (cm)'], data['petal width (cm)'], s = 100,c=data['target'], marker='s', cmap=cm.jet)
           
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X,y)
```

```python
knn.predict([3,5,4,2])
x_new = [[3,5,4,2],[5,4,3,2]]
knn.predict(x_new)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X,y)
knn.predict(x_new)
```

```python
from sklearn import metrics
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state = 4)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print(metrics.accuracy_score(y_test, y_pred))
```

```python
knn = KNeighborsClassifier(n_neighbors = 1)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print(metrics.accuracy_score(y_test, y_pred))
```

```python
k_range = range(1, 31)
scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    scores.append(metrics.accuracy_score(y_test, y_pred))
```

```python
plt.plot(k_range, scores, color="green")
plt.xlabel("K-Value")
plt.ylabel("Accuracy Score") 
```
